```python
import pandas as pd
#df['Labels'] = label_data(df['Absorbance'])
import pandas as pd

#df = pd.read_csv("drug_candidate1.csv", sep=';')
df = pd.read_table("/drug_candidate.txt")
df.head()



def label_data(absorbance):
    if absorbance <= 0.35:
        category_label = "LL"
    elif 0.35 >= absorbance <= 0.54:
        category_label = "HH"
    elif 0.55 <= absorbance <= 0.72:
        category_label = "LH"
    elif 0.73 <= absorbance <= 0.9:
        category_label = "HL"
    else:
        category_label = "Cannot be placed"
    return category_label

# Apply the label_data function to create a new column 'Labels'
df['Labels'] = df['Absorbance'].apply(label_data)

# Print the updated DataFrame
print(df)
```
